```python
import pandas as pd
import numpy as np
```

```python
AC = pd.read_json('ActualCoordination.json')
```

```python
AC.shape
```

```python
AC.describe()
```

```python
TNM_CR = pd.read_json('CoordinationNeeds') # calculated using TNM tool.
```

```python
TNM_CR_NORM = TNM_CR / TNM_CR.max(axis=0)
pd.DataFrame(TNM_CR_NORM).describe()
```

```python
TA = pd.read_json('TaskAssignment.json')
TD = pd.read_json('TaskDependency.json')
```

```python
TA.shape
```

```python
TD.shape
```

```python
#CR = np.dot(np.dot(TA, TD),  TA.T)
CR = TA @ TD @ TA.T   
```

```python
CR.shape
```

```python
CR_normalised = CR / CR.max(axis=0)
pd.DataFrame(CR_normalised).describe()
```

```python
diff =  (CR_normalised - AC)
```

```python
pd.DataFrame(diff).describe()
```

### Calculation of CR using TA* TD *TA.T

```python
CR_count=0
for i in range(0, len(CR_normalised)):
    for j in range(len(CR_normalised[0])):
        if CR_normalised[i][j] > 0:
            CR_count+=1
print(CR_count)
```

```python
diff_count=0
for i in range(0, len(diff)):
    for j in range(len(diff[0])):
        if diff[i][j] > 0:
            diff_count+=1
print(diff_count)
```

```python
congruence = diff_count / CR_count
print(congruence)   ## CR is calculated using TA * TD * TA.T
```

### Calculation of CR through tnm tool

```python
tnm_cr_count=0
for i in range(0, len(TNM_CR_NORM)):
    for j in range(len(TNM_CR_NORM[0])):
        if TNM_CR_NORM[i][j] > 0:
            tnm_cr_count+=1
print(tnm_cr_count)
```

```python
diff = TNM_CR_NORM - AC
```

```python
diff_count=0
for i in range(0, len(diff)):
    for j in range(len(diff[0])):
        if diff[i][j] > 0:
            diff_count+=1
print(diff_count)
```

```python
congruence = diff_count / tnm_cr_count  
print(congruence) ## CR is calculated using TNM tool.
```
